### _Question 1_

- _Define a bond and explain how it is issued_
- _Explain the relationship between bonds and interest rates_

A bond is a deft issues by a companies or government for a defined period of time at a fixed interest rate. For example, bonds are use by the government to finance projects.

By buying the bond, the bondholders lend to monet to the bond issuer. The issuer promise to pay back the borrowed amount (face value or principal), at a specific future date (maturity date)

During this period, the bond issuer pays periodic interest payments (coupon) for which the amound doesn't change over the life bond to the bondholder until the bond reaches maturity.

Bonds are issued through a process called underwriting. Here's how it generally works:

1. **Issuer Decides to Borrow:** The entity in need of funds, such as a corporation or government, decides to borrow money by issuing bonds to finance projects.
2. **Selection of Terms:** The issuer determines the terms of the bond, including the principal amount, interest rate, maturity date, and frequency of interest payments.
3. **Underwriting:** The issuer typically hires an investment bank or underwriting syndicate to help sell the bonds to investors. The underwriter assesses the creditworthiness of the issuer and helps determine an appropriate interest rate for the bonds.
4. **Bond Offering:** The bonds are offered to investors through a bond offering. Investors can purchase the bonds directly from the issuer or through the underwriter.
5. **Bond Sale:** Investors buy the bonds, providing the issuer with the funds it needs. In return, the issuer promises to repay the principal amount at maturity and make periodic interest payments.
6. **Trading on Secondary Market:** After the initial sale, bonds can be traded on the secondary market between investors. The price of a bond on the secondary market can fluctuate based on changes in interest rates, credit risk, and other factors.

The relationship between bonds and interest rates is inverse. When interest rates rise, newly issued bonds typically offer higher yields to attract investors and because it's riskier when the interest rate grow up. This means that existing bonds with lower interest rates become less attractive in comparison, causing their prices to decrease in the secondary market. Conversely, when interest rates fall, newly issued bonds tend to offer lower yields, making existing bonds with higher interest rates more valuable, leading to an increase in their prices. This relationship between bond prices and interest rates is known as the interest rate risk. In other words, the price of a bond changes based on the economic conditions due to the fact that the interest rates change also according to the economic conditions.

The main advantage of bonds is that it offer security of stable interest payments

The spot rate corresponds to the interest rates with a specific maturity period for which a person can borrow funds today. The sport rate is also known as the yield to maturity for a zero-coupon bond. As a zero coupon bond does not pay coupons, it is sold at a discount from its par value.

The main characteristics of spot rates are:
- **Scpecific maturity:** Each spot rate corresponds to a particular maturity period. Hence, each spot rate is different for each maturity 
- **Instantaneous rate:** Spot rates represent the interest rate that would be earned on an investment made immediately, without any intermediate cash flows.
- **Forward-looking:** Despite being a current rate, spot rates are forward-looking because they indicate the market's expectation of future interest rates for each maturity period.

Spot rates are derived from the market through a process called **Bootstrapping**.

**1. Obtain Market Prices:** Gather prices for various fixed-income securities with different maturity periods. These securities could include government bonds, corporate bonds, or other fixed-income instruments.
**2. Identify Zero-Coupon Bonds:** Zero-coupon bonds are bonds that do not make periodic interest payments but are sold at a discount to face value and pay the face value at maturity. By identifying the prices of these bonds, we can derive the spot rates for each maturity period.

To derive the spot rates from zero-coupon bonds, we employ the following equation :

$y = m (\frac{M}{P}^{\frac{1}{m*n}}-1]$

Where:

* $y$: the spot rate 
* $P$: the present value of a zero coupon bond (or a STRIP) 
* $M$: the par value of the STRIP
* $n$: number of years 
* $m$: number of periods per year


**3. Use Yield Curve:** Once we have the prices of zero-coupon bonds with different maturities, we can use them to calculate the spot rates for each maturity. The process involves solving for the interest rate that equates the present value of each bond's future cash flows (the face value at maturity) to its current market price.
**4. Interpolate:** Sometimes, market data may not be available for every maturity period. In such cases, spot rates for intermediate maturities can be estimated using interpolation techniques.

To derive the spot rates from zero-coupon bonds, we employ the following equation :

$y = m [(\frac{M}{PV})^{\frac{1}{m*n}}-1]$

Where:

* $y$: the spot rate 
* $PV$: the present value of a zero coupon bond (or a STRIP) 
* $M$: the par value of the STRIP
* $n$: number of years 
* $m$: number of periods per year

In fact, derive the price of a zero-coupon bond relies on compute the _Present Value (PV)_

Hence, we can rearrange the following formula to solve for the present value:

$$
PV = \frac{M}{{(1 + \frac{y}{m}})^{m*n}}
$$



```python
# Par value of the STRIP
par_value = 100

# Number of years until maturity
n = [0.5, 1.0, 1.5, 2.0, 2.5, 3.0, 3.5, 4.0, 4.5, 5.0]

# Number of periods per year (semi-annual)
m = 2

# Calculate the price of each STRIP
strip_prices = []
for i in range(len(spot_rates)-1):
    price = par_value / ((1 + (spot_rates[i] / 100) / m) ** (m * n[i]))
    strip_prices.append(price)

# Check if STRIPS are below or above par value
for i, price in enumerate(strip_prices):
    if price < par_value:
        print(f"STRIP {n[i]} years is below par value")
    elif price > par_value:
        print(f"STRIP {n[i]} years is above par value")
    else:
        print(f"STRIP {n[i]} is at par value")

# Print the calculated prices of the STRIPS
print("\nPrices of the STRIPS:")
for i, price in enumerate(strip_prices):
    print(f"STRIP {n[i]} years: ${price:.2f}")
```

#### Question 6:

- Define a forward rate and its characteristics. Explain how forward rates are derived from the market. 
- From the chosen spot rates, calculate the respective forward rates. - Plot the forward yield curve


The forward rate represents the interest rate of a future period. For instance, if a person wants to borrow money to buy a house in one year, he will borrow at the rate that will be applied one year from today, i.e the one year forward rate

The characteristics of forward rate are:

* **Forward rates are forward-looking:** They provide information about future expectations regarding interest rates, exchange rates, or commodity prices.
* **Based on market expectations:** Forward rates are derived from market expectations of future economic conditions, supply and demand dynamics, and other relevant factors.
* **Arbitrage-free:** Forward rates should reflect market equilibrium and should not allow for riskless arbitrage opportunities. In efficient markets, forward rates adjust to eliminate arbitrage opportunities.
* Investors are indifferent between an investment with a spot rate or with a forward rate as the return is the same
* The forward yield curve is the graph that shows the relationship between forward rates and maturity

Forward rate are derived from spot rates. The process involves using the yield curve, which represents the relationship between interest rates and time to maturity. Forward rates can be calculated using the current spot rates and interest rate differentials for the desired forward period.

To deduct a forward rate $f(t)$ the following equation is applied:

$$
(1+\frac{s_{t}}{m})^{l_{s_{t}}}=(1+\frac{s_{t-1}}{m})^{l_{s_{t-1}}}\times(1+\frac{f(t)}{m})^{1}
$$

where:

* $s_t:$ spot rate at time $t$ (example if $t 1.5$ years and if semi annual periods, $s_t$ will be the spot rate for year 1.5
* $s_{t-1}$ spot rate at time $t-1$ (example if $t=1.5$ years and if semi annual periods, $s_{t-1}$ will be the spot rate for year $1$
* $l$ number of periods (example if $t=1.5$ years and if semi annual periods, $l=3$ as $3$ semi annual periods in $1.5$ years
* $m$ number of periods per year (example if semi annual periods, $m=2$ as $2$ semi annual periods in one year)
* $f(t)$ the forward rate corresponding to the spot rate $s_t$ (example if $s_t$ is the spot rate for $1.5$ year investment, $f(t)$ will be the forward rate for an investment between year $1$ and year $1.5$ following a first $1$ year investment with the spot rate $s_{t-1}$

$$
f(t) = \left( \left(1 + \frac{s_t}{100 \times m} \right)^{l_{s_t}} \middle/ \left(1 + \frac{s_{t-1}}{100 \times m} \right)^{l_{s_{t-1}}} \right) - 100 \times m
$$

Then we multiply $f(t)$ by the term $\times 100 \times m$ to convert the calculated forward rate from a decimal to a percentage. The forward rate is typically expressed as an annualized rate, and multiplying by 100 converts the decimal to a percentage. The $m$ term represents the number of periods per year, so multiplying by $m$ adjusts the forward rate for the actual number of periods in a year.

```python
def calculate_forward_rate(spot_rate_t, spot_rate_t_1, m, l_s_t, l_s_t_1):
    return (((1 + spot_rate_t / (100 * m)) ** l_s_t) / ((1 + spot_rate_t_1 / (100 * m)) ** l_s_t_1) - 1) * 100 * m

# Number of periods per year (semi-annual)
m = 2

# Calculate the respective forward rates
forward_rates = []
for i in range(1, len(spot_rates)):
    spot_rate_t = spot_rates[i]
    spot_rate_t_1 = spot_rates[i - 1]
    l_s_t = i  # Number of semi-annual periods from year 0 to year t
    l_s_t_1 = i - 1  # Number of semi-annual periods from year 0 to year t-1
    forward_rate = calculate_forward_rate(spot_rate_t, spot_rate_t_1, m, l_s_t, l_s_t_1)
    forward_rates.append(forward_rate)

# Print the calculated forward rates
print("Calculated Forward Rates:")
for i, rate in enumerate(forward_rates):
    print(f"Forward Rate  f({i + 0.5}): {rate:.4f} %")
    
# Plot the calculated forward rates
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(spot_rates)), forward_rates, marker='o', linestyle='-')
plt.title('Forward Rates')
plt.xlabel('Period')
plt.ylabel('Forward Rate (%)')
plt.grid(True)
plt.xticks(range(1, len(spot_rates)))
plt.show()
```

To price a bond using the interest rates (i.e. spot rate) we can invoke the following.

$$

$$

where:
* $P$ (Present value): The value (price) of the bond at the day of estimation. It takes into account all future cash flows (coupons + principal).
* $M$ (Par value or nominal or face value or principal amount): This is the amount shown on a bond certificate that bond purchaser (bond holder/lender) lends to a bond isuer (borrower). This amount is repaid to the lender at maturity. The par value of one bond is typically \$100,\$1000 or \$10000. In addition to the par value, the bond purchaser may also receive periodic interest payment 
* $C$: A coupon is a fixed return that a bond holder earns periodically until the bond matures. It is calculated as a nominal percentage of the par value of the bond.
* $y$ (Discount rate or yield): Interest rate at which future returns are discounted in order to estimate their present value. The disocunting method is based on the principle that money received in the future is not worth as much as an equal sum received today.
* $N$ (maturity): Bonds have a number of periods to maturity $m$. These are tipically annual periods but may also be semi-annual or quarterly. The number of periods equals the number of coupon payments.

### Question 8
Assume that you have purchased the 4 year government bond 60 days after it was issued. Using the spot rates, calculate the dirty and the clean price of the bond

**Fractional period compounding:** 

To estimate the dirty price of a bond taking into account fractional periods, the following equation is applied:
$$
P = \frac{C}{(1+\frac{y}{m})^{w}} + \frac{C}{(1+\frac{y}{m})^{1+w}} + \dots + \frac{C}{(1+\frac{Z}{m})^{N-1+w}} + \frac{M}{(1+\frac{y}{m})^{N-1+w}} = \sum_{i=0}^{N-1} \frac{C}{(1+\frac{y}{m})^{i+w}} + \frac{M}{(1+\frac{y}{m})^{N-1+w}}
$$

where:

* $P$: price of the bond (present value)
* $C$: coupon
* $Y$: yield (or discount rate or interest rate)
* $N$: Maturity period
* $m$: number of periods per year
* $M$: par value of the bond (the principal amount)
* $w$: the number of days until the next coupon payment divided by the number of days in the coupon period (The number of days to use for the coupon period is determined by the appropriate day count convention. For example, the denominator is 180 for semi-annual bonds that use the 30/360 convention.).

**For example (computing the dirty price of a bond):**

Suppose the bond from the previous example (10% coupon bond annually, with par value = $1,000) is a U.S. corporate bond that pays coupons semi-annually on January 1 and July 1. Assume that today is April 1, 2015, and the bond matures on July 1, 2025. Compute the dirty price of this bond if the required annual yield is 8%.

**Anwser:**
This is a U.S. corporate bond, so it uses the 30/360 day count convention. Under this convention the number of days between today (April 1, 2015) and the next coupon payment (July 1, 2015) is 90 days (3 months at 30 days per month). Therefore, $C=\frac{0.1 * 1000}{2}=50$ and $w=\frac{90}{180}=0.5$ since $N = 21 \implies N - 1 + w = 20.5$. Therefore, the dirty price of
the bond is:

$$
P = \sum_{i=0}^{N-1} \frac{C}{(1+\frac{y}{m})^{i+w}} + \frac{M}{(1+\frac{y}{m})^{N-1+w}} = \sum_{i=0}^{20} \frac{50}{(1+\frac{0.08}{2})^{i+0.5}} + \frac{1000}{(1+\frac{0.08}{2})^{20.5}} = \$1,162.87
$$

Please note that here $N=21$ for the following reasons:
Since the bond pays semi-annual coupons, the remaining period from April 1, 2015, to July 1, 2025, includes both full and partial coupon periods.

The total number of semi-annual periods can be calculated as follows:

* 10 full years contribute 20 semi-annual periods.
* The partial period from April 1 to July 1, 2025, is less than a year, so it contributes a fraction of a semi-annual period.
In this case, the partial period contributes half of a semi-annual period because it's three months, which is half of a year.

Therefore, the total number of semi-annual periods (N) is the sum of the full periods and the fractional period:

$N = 20 \text{(full periods)} + 0.5 \text{(fractional period)} = 20.5$

This is why $N$ is equal to $21$

To compute the *clean price* we have first to compute the *Accrued Interest (AI)*. When a bond is purchased not on a coupon date, the buyer of the bond must pay its owner any interest earned from the day of the last coupon payment up to the settlement date (= purchase date):

To compute the Accrued Interest (AI), we can refer to the following equation: 

$$
AI = c \times (\frac{\text{number of days from last coupon to the settlement date}}{\text{number of days in a coupon period}})
$$

where:
* $c$: coupon payment

**Example: computing AI**

A $1,000 par value U.S. corporate bond pays a semi-annual 10% coupon. An investor purchases the bond. However, assume the last coupon was paid 90 days ago and there are 30 days in each month. What is the accrued interest the bond buyer must pay to the bond seller?

**Answer:**
$$
c = \frac{0.1*1000}{2} = 50 \implies AI = 50 (\frac{90}{180}) = \$25
$$

Finally to compute the *clean price* we can use the following formula:

$$
\text{clean price} = \text{dirty price} - \text{accrued interest}
$$


#### In our case:

We assume that is a U.S. corporate bond that pays semi-annualy (i.e $m=2$) a 4% coupon, so it uses the 30/360 day count convention. Under this convention the number of days between today  (1st of january 2024) and the next coupon payment is 60 days with a face value of 1000 dollars. Therefore, $C=\frac{0.04 * 1000}{2}=20$ and $w=\frac{60}{180}=0.33$ since $N = 4 \times 2 = 8 \implies N - 1 + w = 7$. Therefore, the dirty price of
the bond is:

$$
P = \sum_{i=0}^{N-1} \frac{C}{(1+\frac{y}{m})^{i+w}} + \frac{M}{(1+\frac{y}{m})^{N-1+w}} = \sum_{i=0}^{20} \frac{50}{(1+\frac{0.08}{2})^{i+0.5}} + \frac{1000}{(1+\frac{0.08}{2})^{20.5}} = \$1,162.87
$$

```python
# Par value of the bond
M = 1000

# Coupon rate (assume annual coupon rate of 4%, so semi-annual rate is 2%)
annual_coupon_rate = 0.04
semi_annual_coupon_rate = annual_coupon_rate / 2
C = semi_annual_coupon_rate * M

# Number of periods per year (semi-annual payments)
m = 2

# Maturity of the bond in semi-annual periods (4 years, so 8 semi-annual periods)
N = 4 * m

# Number of days since the last coupon payment (60 days after issuance)
days_since_last_coupon = 60

# Calculate the fraction of the coupon period until the next coupon payment (w)
days_in_coupon_period = 180  # Assuming 30/360 day count convention
w = days_since_last_coupon / days_in_coupon_period

# Calculate the present value of the bond using spot rates and fractional period compounding
present_value = sum([C / ((1 + spot_rate / (100 * m)) ** (i + w)) for i, spot_rate in enumerate(spot_rates[:N])]) + M / ((1 + spot_rates[N-1] / (100 * m)) ** (N-1 + w))

# Calculate the accrued interest
accrued_interest = C * (days_since_last_coupon / days_in_coupon_period)

# Calculate the dirty price
dirty_price = present_value

# Calculate the clean price
clean_price = dirty_price - accrued_interest

print("Accrued interest:", accrued_interest)
print("Dirty price of the bond:", dirty_price)
print("Clean price of the bond:", clean_price)
```

### Question 9:

- Estimate the yield to maturity (YTM) of the bond calculated in question 7 a) Interpret

Yield to maturity (YTM) is the realized return of a bond, assuming it is held until maturity. It is expressed as an annual rate.

$$
P = \sum_{i=1}^{N} \frac{C}{(1 + \frac{YTM}{m})^i} + \frac{M}{(1 + (\frac{YTM}{m}))^N}
$$

where:

* $P$: price (present value) of a bond
* $C$: Coupon of the bond
* $YTM$: The yield to maturity
* $N$: Maturity period
* $M$: par value of the bond (the principal amount)
* $m$: number of periods per year

**Example: Spot rates and YTM**
A bond with a \$100 par value pays a 5% coupon annually for 4 years. The spot rates corresponding to the payment dates
are as follows: 

* Year 1: 4.0%
* Year 2: 4.5%

Let's proceed with the calculation of YTM using these steps. We may need to employ numerical methods to solve for YTM, as it's typically not possible to find a closed-form solution.

The function we will estimate the YTM by using optimize library of scipy in order to minimize the difference between the true price of the bond and the estimated price of the bond.

```python
from scipy.optimize import minimize

# Bond characteristics
M = 1000  # Par value
C = 20    # Coupon payment (semi-annual)
N = 8     # Number of periods until maturity
P = 977.19  # Price of the bond

# Define the function to calculate the present value of the bond given YTM
def present_value(YTM):
    PV = sum([C / ((1 + YTM/2) ** i) for i in range(1, N+1)]) + M / ((1 + YTM/2) ** N)
    return PV

# Define the objective function for optimization
def objective_function(YTM):
    return abs(P - present_value(YTM))

# Initial guess for YTM
initial_guess = 0.05  # 5% initial guess

# Optimize the objective function to find the YTM
result = minimize(objective_function, initial_guess, bounds=[(0, 1)])

# Extract the YTM from the optimization result
YTM = result.x[0] * 100

print("The estimated YTM: " + str(YTM) + "%")
```

#### a) Calculate DV01 for a five basis points decrease in YTM Interpret

DV01 - “dollar value of an 01”, measures the change in a fixed income security’s value in $ for every one basis point change in interest rates. The “01” refers to one basis point (= 0.01% = 0.0001). It is a good measure for relatively small changes in interest rates. Large changes do not follow a linear function, so the DV01 estimate would be incorrect.

DV01 is calculated using the following formula:

$$
DV01 = - \frac{\Delta BV}{10000 \times \Delta y}
$$

where: 

* $\Delta BV$ = change in value of a bond price
* $\Delta y$ = change in yield (YTM)

**Example: Computing DV01**

Suppose the yield on a zero-coupon bond declines from 3.00% to 2.99%. As a result, the price of the bond increases from $\$17.62$ to $\$17.71$. 
Compute the DV01: 

$$
DV01 = - \frac{\Delta BV}{10000 \times \Delta y} = - \frac{\$17.71 - \$17.62}{10000 \times (-0.0001)} = \frac{\$0.09}{1} = \$0.09
$$

Interpretation: The value of the bond will change by $0.09 for every single basis point change in the yield of the bond

In our case we will a five basis points decrease in YTM means that $\Delta y = -0.0005$ and according to the 
$$
DV01 = - \frac{\Delta BV}{10000 \times \Delta y}
$$

#### b) Calculate duration and convexity for a 0 75 increase and decrease in YTM Interpret

**Duration**
Duration measures the change in a fixed income security’s value in % for every 1% change in interest rates. It is a good measure for relatively small changes in interest rates. Large changes do not follow a linear function, so the duration estimate would be incorrect.

The most used method of estimating duration is called ”effective duration". It is calculated using the following formula:

$$
\text{effective duration} = D = \frac{BV_{-\Delta y} - BV_{+\Delta y}}{2 \times BV_{0} \times \Delta y}
$$

where: 
* $BV_{-\Delta y} = $ estimated price of the bond if the yield decreases by a given number of basis points
* $BV_{+\Delta y} = $ estimated price of the bond if the yield increases by a given number of basis points
* $BV_{0} = $ initial observed bond price
* $\Delta y = $ change in yield

**Example: Computing duration**
Suppose there is a 15-year bond with an annual coupon of 7% trading at par value (=$100) and a yield of 7%. Compute and interpret the bond’s duration for a 50 basis points (0.50%) increase and decrease in yield.

* If interest rates rise by 50 basis points (yield = 7.5%), the estimated price of the bond will fall to $95.59
* If interest rates fall by 50 basis points (yield = 6.5%), the estimated price of the bond will increase to $104.71.

Therefore, the duration of the bond is:

$$
\text{effective duration} = D = \frac{BV_{-\Delta y} - BV_{+\Delta y}}{2 \times BV_{0} \times \Delta y} = \frac{104.71 - 95.59}{2 \times 100 \times 0.005} = 9.115%
$$

**Interpretation:**
For a 100 basis point (1%) change in yield, the expected price change is 9.115%. In other words, if the yield on this bond increases by 1%, the price should decrease by about 9.115%. If the yield decreases by 1%, the price of the bond should increase by about 9.115%.

**Convexity**

Convexity is a measure of the curvature of the relationship between bond yield and price. It is good measure for relatively large changes in interest rates.
* "DV01 and “duration” are linear estimates. They assume that the change in the price of a bond is proportional to the change in interest rates. However, the greater the change in the respective interest rate, the greater the curvature (= the slope) between the bond price and the yield will be. As convexity increases, the systemic risk to which a portfolio is exposed increases.
$$
\text{convexity} = Cx = \frac{BV_{-\Delta y} + BV_{+\Delta y} - 2 \times BV_{0}}{BV_{0} \times \Delta y^2} = \frac{104.71 - 95.59}{2 \times 100 \times 0.005} = 9.115%
$$

where: 
* $BV_{-\Delta y} = $ estimated price of the bond if the yield decreases by a given number of basis points
* $BV_{+\Delta y} = $ estimated price of the bond if the yield increases by a given number of basis points
* $BV_{0} = $ initial observed bond price
* $\Delta y = $ change in yield

**Example: Computing convexity**
Suppose there is a 15-year bond with an annual coupon of 7% trading at par value (=$100) and a yield of 7%.

* If interest rates rise by 50 basis points (0.50%), the estimated price of the bond is $95.59
* If interest rates fall by 50 basis points (0.50%), the estimated price of the bond is $104.71.
$$
\text{convexity} = Cx = \frac{BV_{-\Delta y} + BV_{+\Delta y} - 2 \times BV_{0}}{BV_{0} \times \Delta y^2} = \frac{104.71 - 95.59 - 2 \times 100}{100 \times 0.005^2} = 114.8
$$

```python
# Initial bond parameters
M = 1000  # Par value
C = 20    # Coupon payment (semi-annual)
N = 8     # Number of periods until maturity
YTM_initial = 4.63125803048071 / 100  # Initial YTM

# Change in YTM for increase and decrease
delta_y = 0.0075

# Compute the decreased YTM
YTM_decreased = YTM_initial - 0.0075 # five basis point decrease
print("New YTM for a 0.75% decreased in original YTM:", YTM_decreased)

# Compute the increased YTM
YTM_increased = YTM_initial + 0.0075 # five basis point increased
print("New YTM for a 0.75% increased in original YTM:", YTM_increased)

# Compute bond price with decreased YTM
P_decreased = sum([C / (1 + YTM_increased / 2) ** i for i in range(1, N + 1)]) + M / (1 + YTM_increased / 2) ** N

# Compute bond price with increased YTM
P_increased = sum([C / (1 + YTM_decreased / 2) ** i for i in range(1, N + 1)]) + M / (1 + YTM_decreased / 2) ** N

# Compute the duration
duration = (P_increased - P_decreased) / (2 * M * delta_y)

# Compute the convexity
convexity = (P_decreased + P_increased - 2 * M) / (M * delta_y ** 2)

# Print the results
print("New price for a 0.75% increase in YTM:", P_decreased)
print("New price for a 0.75% decrease in YTM:", P_increased)
print(f"Duration for a 0.75% increase and decrease in YTM: {duration:.4f}")
print(f"Convexity for a 0.75% increase and decrease in YTM: {-convexity:.4f}")
```

Regarding the convexity, unlike “DV01” or ”duration”, a convexity of 795.10 cannot be conveniently converted into some measure of potential price volatility. The convexity value means nothing in isolation. However, it becomes very useful when it is used to measure a bond’s convexity effect.

A bond’s convexity effect, combined with a bond’s duration effect, provides a more accurate estimate of a potential price change as it takes into account both small and big changes in interest rates.

#### c) Create a hedging strategy with the two bonds using the duration/convexity approach and Interpret
This is a process in which both measures must be hedged simultaneously. To do this, the investor must use two hedging instruments (two bonds) - one per measure. To create a perfect hedge, an investor should know the ”duration” and the “convexity” of each of the hedging instruments. The following two equations must be satisfied for a position to be hedged.

\[
\begin{align*}
    V_{OP} \times D_{OP} - V_{HI1} * D_{HI1} - V_{HI2} \times D_{HI2} = 0 \\
    V_{OP} \times Cx_{OP} - V_{HI1} * Cx_{HI1} - V_{HI2} \times Cx_{HI2} = 0
\end{align*}
\]

where:

* $D_{OP}=$ duration of the investor’s current open position
* $V_{OP}=$ total value of the investor’s current open position
* $Cx_{OP}=$ convexity of the open position
* $D_{HI}=$ duration of the hedging instrument used to protect the current open position
* $Cx_{HI}=$ convexity of the hedging instrument used to protect the current open position

We need to solve this system of equations with 2 unknowns to find out the value of the hedge.

For reminder the sensitivity measures we found for the previous question are the following:
* Duration = 3.5653 
* Cx (convexity) = 795.7091
* Value = 5.5 millions dollars

We use these sensitivity measures for the open position.

Regarding the convexity and the duration of the two bonds to create a hedging strategies, we used the value that we generated randomly previously

```python
from scipy.optimize import fsolve

# Given sensitivity measures for the open position
duration_OP = 3.5653
convexity_OP = 795.7091
value_OP = 5_500_000  # in millions of dollars


# Define the system of equations
def equations(vars):
    V_HI_1, V_HI_2 = vars
    eq1 = value_OP * duration_OP - V_HI_1 * duration_HI_1 - V_HI_2 * duration_HI_2
    eq2 = value_OP * convexity_OP - V_HI_1 * convexity_HI_1 - V_HI_2 * convexity_HI_2
    return [eq1, eq2]

# Solve the system of equations
initial_guess = [0, 0]  # Initial guess for the values of the bonds
V_HI_1, V_HI_2 = fsolve(equations, initial_guess)

# Print the results
print("Value of bond 1 (in million):", str(V_HI_1))
print("Value of bond 2 (in million):", str(V_HI_2))
```

* Value of bond 2 (in million): 94.19
    - Similarly, we would need approximately $94.19 million worth of bond 2 to hedge your open position.
